```python
from manim import *
```

```python
from datetime import datetime
import numpy as np

num_dias = np.arange(1, 8, 1) # Los días del 1 al 7
dias = ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes", "Sabado", "Domingo"]
dias_dict = dict(zip(num_dias, dias))

print(dias_dict[datetime.today().weekday()])
```

```python
%%manim -ql -v WARNING Bienvenidos


class Bienvenidos(Scene):
  def construct(self):
    text1 = Tex(r"Manim en Colab :)")
    text2 = Tex(f"@mnsosa les desea un feliz {dias_dict[datetime.today().weekday()]}")
    
    self.play(Write(text1))
    self.wait()

    self.play(Transform(text1, text2))
    self.wait()
```
